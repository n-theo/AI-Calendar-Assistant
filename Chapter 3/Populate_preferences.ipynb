```python
def create_week(week_len, num_slots):
  week_array = np.zeros((week_len, num_slots))
  return week_array
```

```python
def get_zeros(lst, min, max):
  return [i+min for i, e in enumerate(lst[min:max]) if e == 0]
```

```python
def num_pieces(num,lenght, max):
    number = num
    ot = list(range(1,lenght+1))[::-1]
    all_list = []
    for i in range(lenght-1):
      n = 1000
      while n > max:
        n = rd.randint(1, num-ot[i])
      all_list.append(n)
      num -= n
    all_list.append(num) 
    if all_list[-1] > max:
       num_pieces(number,lenght, max)

    return all_list
```

```python
def imb_week(num_days,  late):
  week = [0,0,0,0,0,0,0]
  lst = []
  for i in range(num_days):
    lst.append(rd.randint(i+1,i+2))
  if not late:
    lst = sorted(lst, reverse=True)
  week[len(week)-num_days-1:num_days+1] = lst
  
  return week
    
```

```python
def label_row(row):
  tuple = [None,None,None]
  if row['Week'] == None:
    tuple[0] = 0
  if row['Week'] == 'early':
    tuple[0] = 1
  if row['Week'] == 'late':
    tuple[0] = 2
  if row['Day'] == None:
    tuple[1] = 0
  if row['Day'] == 'early':
    tuple[1] = 1
  if row['Day'] == 'late':
    tuple[1] = 2
  if row['b2b'] == True:
    tuple[2] = 0
  if row['b2b'] == False:
    tuple[2] = 1
  return tuple 

```

```python
df['label'] = df.apply(lambda row: label_row(row), axis=1)
```
